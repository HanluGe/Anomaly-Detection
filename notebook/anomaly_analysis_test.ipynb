```python
def detect_price_jumps_lastpx_rolling5days(
    df,
    minutes_per_day=265,
    window_days=5,
    threshold=3
):
    """
    ÂØπÂçïÂè™ËÇ°Á•®ÁöÑ minutely_df Ê£ÄÊµã jump
    Rolling window = window_days * minutes_per_day
    """
    
    df = df.copy()

    # 1) ÂçïÊó• diffÔºà‰∏ç‰ºöË∑®Êó•Ôºâ
    df["px_ret"] = df.groupby("date")["lastPx"].diff()

    # 2) rolling ËÆ°ÁÆó 5 Â§© std
    window = minutes_per_day * window_days
    df["sigma"] = (
        df["px_ret"]
        .rolling(window=window, min_periods=window//2)
        .std()
    )

    df["sigma"] = df["sigma"].replace(0, np.nan)

    # 3) z-score
    df["zscore"] = df["px_ret"] / df["sigma"]

    # 4) Âà§ÂÆö jump
    df["jump"] = df["zscore"].abs() > threshold

    return df

def batch_apply_detector(
    stock_data_dict,
    detector_fn,
    df_key="minutely_df",
    verbose=True,
    **detector_kwargs,
):

    result_dict = {}

    for stock, data in stock_data_dict.items():
        if df_key not in data:
            if verbose:
                print(f"[Warning] {stock} Ê≤°Êúâ {df_key}ÔºåË∑≥Ëøá")
            continue

        if verbose:
            print(f"Applying detector for {stock} ...")

        df = data[df_key]
        result_df = detector_fn(df, **detector_kwargs)
        result_dict[stock] = result_df

    if verbose:
        print("Detection finished for:", list(result_dict.keys()))

    return result_dict
```

```python
def intraday_event_count(df, col_name, bin_size=5,
                         start_time="09:00", end_time="13:25"):
    """
    ÁªüËÆ° df[col_name]==True ÁöÑ‰∫ã‰ª∂ÔºåÂú®‰∏ÄÂ§©ÂÜÖÁöÑÊó∂Èó¥ÂàÜÂ∏É„ÄÇ
    """
    df = df.copy()
    df["minute"] = df.index.hour * 60 + df.index.minute  # ËΩ¨ÂàÜÈíü

    # Âè™ÁªüËÆ°ÂºÇÂ∏∏‰∫ã‰ª∂
    event_minutes = df[df[col_name]]["minute"]

    # Âå∫Èó¥ÂÆö‰πâ
    h1, m1 = map(int, start_time.split(":"))
    h2, m2 = map(int, end_time.split(":"))

    start_min = h1 * 60 + m1
    end_min   = h2 * 60 + m2

    bins = np.arange(start_min, end_min + 1, bin_size)

    # ÁªüËÆ°‰∫ã‰ª∂Ê¨°Êï∞
    counts, edges = np.histogram(event_minutes, bins=bins)

    # ËøîÂõûÊ†ºÂºèÂåñÂêéÁöÑ Series
    index = pd.Index(edges[:-1], name="minute_bin")
    counts_series = pd.Series(counts, index=index, name=f"{col_name}_count")

    return counts_series
```

```python
def plot_intraday_event_hist_multi_pretty(
    df_dict,
    col_name,
    bin_size=5,
    start_time="09:00",
    end_time="13:25",
    stock_list=None,
):
    """
    Êõ¥ÁæéËßÇÁöÑÁâàÊú¨ÔºöÊ®™ÊéíÂ§ö‰∏™ËÇ°Á•®ÁöÑÊó•ÂÜÖ‰∫ã‰ª∂ÂàÜÂ∏ÉÂõæ
    """

    # ÈªòËÆ§ÂèñÂâç 3 Âè™ËÇ°Á•®
    all_stocks = list(df_dict.keys())
    if stock_list is None:
        stock_list = sorted(all_stocks)[:3]
    else:
        stock_list = [s for s in stock_list if s in df_dict]

    n = len(stock_list)
    if n == 0:
        print("‚ö†Ô∏è No valid stocks to plot.")
        return

    # Êó∂Èó¥Â§ÑÁêÜ
    h1, m1 = map(int, start_time.split(":"))
    h2, m2 = map(int, end_time.split(":"))
    start_min = h1 * 60 + m1
    end_min   = h2 * 60 + m2

    xticks = np.arange(start_min, end_min + 1, 30)
    xtick_labels = [f"{h:02d}:{m:02d}" for h, m in [(t//60, t%60) for t in xticks]]

    # ---------------------------------------------
    # üé® Ê†∑Âºè
    # ---------------------------------------------
    text_color = "#3d405b"   # Ê∑±ËìùÁÅ∞
    panel_bg = (244/255, 241/255, 222/255, 0.4)  # Â≠êÂõæËÉåÊôØ

    plt.style.use("ggplot")
    fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=True)

    axes = axes.flatten()

    # ‰ªÖÂ≠êÂõæËÉåÊôØÊîπËâ≤Ôºàfig ËÉåÊôØ‰øùÊåÅÁôΩËâ≤Ôºâ
    for ax in axes:
        ax.set_facecolor(panel_bg)

    colors = ["#e07a5f", "#81b29a", "#f2cc8f"]  # ‰∏âÁßçÈ¢úËâ≤

    # ---------------------------------------------
    # üìä ÁªòÂõæ
    # ---------------------------------------------
    for i, stock in enumerate(stock_list):
        ax = axes[i]
        df = df_dict[stock]

        counts = intraday_event_count(
            df,
            col_name=col_name,
            bin_size=bin_size,
            start_time=start_time,
            end_time=end_time,
        )

        ax.bar(
            counts.index,
            counts.values,
            width=bin_size * 0.9,
            align="edge",
            color=colors[i % len(colors)],
            alpha=0.85,
        )

        # Ê†áÈ¢ò
        ax.set_title(
            f"{stock}",
            fontsize=14,
            fontweight="bold",
            color=text_color,
        )

        ax.set_xlabel("Time of Day", color=text_color)
        if i == 0:
            ax.set_ylabel("Event Count", color=text_color)

        ax.set_xticks(xticks)
        ax.set_xticklabels(
            xtick_labels, rotation=40, ha="right", color=text_color
        )

        # tick È¢úËâ≤
        ax.tick_params(axis="y", colors=text_color)
        ax.tick_params(axis="x", colors=text_color)

        # ËæπÊ°Ü
        ax.spines["top"].set_visible(False)
        ax.spines["right"].set_visible(False)
        ax.spines["left"].set_color(text_color)
        ax.spines["bottom"].set_color(text_color)

        ax.grid(axis="y", alpha=0.25, color=text_color)

    # Âà†Èô§Â§ö‰Ωô ax
    for j in range(n, 3):
        fig.delaxes(axes[j])

    fig.suptitle(
        f"Intraday Distribution of {col_name} (bin={bin_size} min)",
        fontsize=16,
        fontweight="bold",
        color=text_color,
        y=1.05
    )

    fig.tight_layout()
    plt.show()
```

```python
def daily_event_vs_nextday_return(
    df: pd.DataFrame,
    event_col: str,
    price_col: str = "lastPx",
):
    """
    ÂàÜÊûêÔºöÊØèÂ§©ÊüêÁ±ª‚Äú‰∫ã‰ª∂Ê¨°Êï∞‚ÄùÂíå‰∏ã‰∏ÄÂ§©Êó•Êî∂ÁõäÁöÑÂÖ≥Á≥ª„ÄÇ

    ËæìÂÖ•:
        df: ÂøÖÈ°ªÂåÖÂê´
            - price_col  (Â¶Ç 'lastPx')
            - event_col  (bool Êàñ 0/1)
            - 'date'     (Ê≤°ÊúâÁöÑËØù‰ºöËá™Âä®Áî® index.date ÁîüÊàê)
        event_col: ‰∫ã‰ª∂ÂàóÂêçÔºå‰æãÂ¶Ç 'jump'„ÄÅ'spread_abnormal'
        price_col: ‰ª∑Ê†ºÂàóÂêçÔºåÈªòËÆ§ 'lastPx'

    ËøîÂõû:
        daily_df: DataFrameÔºåÂàóÂåÖÊã¨
            - date
            - event_count : ÂΩìÂ§©‰∫ã‰ª∂Ê¨°Êï∞
            - ret_today   : ÂΩìÂ§©ÂºÄÁõòÂà∞Êî∂ÁõòÊî∂Áõä
            - ret_next_day: ‰∏ã‰∏ÄÂ§©Êî∂ÁõäÔºàÂØπÈΩêÂΩìÂâçËøôÂ§©Ôºâ
    """
    df = df.copy()

    if "date" not in df.columns:
        df["date"] = df.index.date

    # ‰øùËØÅÊòØ 0/1
    event_flag = df[event_col].astype(int)

    # Êó•Â∫¶‰∫ã‰ª∂Ê¨°Êï∞
    event_count = event_flag.groupby(df["date"]).sum().rename("event_count")

    # Êó•Â∫¶Êî∂Áõä
    daily_first = df.groupby("date")[price_col].first()
    daily_last = df.groupby("date")[price_col].last()
    ret_today = (daily_last / daily_first - 1).rename("ret_today")

    daily_df = pd.concat([event_count, ret_today], axis=1).reset_index()
    daily_df = daily_df.sort_values("date")

    # ‰∏ã‰∏ÄÂ§©Êî∂Áõä
    daily_df["ret_next_day"] = daily_df["ret_today"].shift(-1)

    return daily_df

def multi_event_ic_table(
    df_dict: dict,
    event_col: str,
    price_col: str = "lastPx",
):
    rows = []

    for stock, df in df_dict.items():
        # ÂÖàÁÆóÂçïËÇ°Á•®ÁöÑÊó•Â∫¶Ë°®
        daily_df = daily_event_vs_nextday_return(
            df, event_col=event_col, price_col=price_col
        )

        # ‰∏¢Êéâ NaNÔºà‰∏ªË¶ÅÊòØÊúÄÂêé‰∏ÄÂ§© ret_next_day ‰∏∫ NaNÔºâ
        tmp = daily_df.dropna(subset=["event_count", "ret_today", "ret_next_day"])

        if tmp.shape[0] < 2:
            ic_today = np.nan
            ic_next  = np.nan
        else:
            ic_today = tmp["event_count"].corr(tmp["ret_today"])
            ic_next  = tmp["event_count"].corr(tmp["ret_next_day"])

        row = {
            "stock": stock,
            "n_days": int(tmp.shape[0]),
            "mean_event_count": float(tmp["event_count"].mean()),
            "mean_ret_today": float(tmp["ret_today"].mean()),
            "mean_ret_next_day": float(tmp["ret_next_day"].mean()),
            "ic_today": float(ic_today) if np.isfinite(ic_today) else np.nan,
            "ic_next": float(ic_next) if np.isfinite(ic_next) else np.nan,
        }
        rows.append(row)

    result_df = pd.DataFrame(rows).set_index("stock")

    # ÂÜçÂä†‰∏ÄË°åÂùáÂÄº
    mean_row = {
        "stock": "MEAN",
        "n_days": np.nan,
        "mean_event_count": result_df["mean_event_count"].mean(),
        "mean_ret_today": result_df["mean_ret_today"].mean(),
        "mean_ret_next_day": result_df["mean_ret_next_day"].mean(),
        "ic_today": result_df["ic_today"].mean(),
        "ic_next": result_df["ic_next"].mean(),
    }
    result_df = pd.concat(
        [result_df, pd.DataFrame([mean_row]).set_index("stock")]
    )

    return result_df
```

```python
def event_forward_path(
    df: pd.DataFrame,
    event_col: str,
    price_col: str = "lastPx",
    horizon: int = 30,
):
    """
    ÂàÜÊûêÔºöÊüêÁ±ª‰∫ã‰ª∂ÂèëÁîü‰πãÂêéÔºåÊú™Êù• horizon ÂàÜÈíü‰ª∑Ê†ºÁöÑÂπ≥ÂùáËµ∞Âäø„ÄÇ

    ËæìÂÖ•:
        df: ÂåÖÂê´
            - event_col (bool / 0/1)
            - price_col
            - date (Ê≤°Êúâ‰ºöËá™Âä®ÁîüÊàê)
        horizon: ÂêëÂêéÁúãÁöÑÂàÜÈíüÊï∞

    ËøîÂõû:
        result: dict
            - offsets   : [0..horizon]
            - paths     : shape=(n_events, horizon+1)
            - mean_path : Âπ≥ÂùáË∑ØÂæÑ
            - n_events  : ‰∫ã‰ª∂‰∏™Êï∞
    """
    df = df.copy()
    if "date" not in df.columns:
        df["date"] = df.index.date

    df[event_col] = df[event_col].astype(bool)

    paths = []
    offsets = np.arange(horizon + 1)

    for d, day_df in df.groupby("date"):
        day_df = day_df.sort_index()
        prices = day_df[price_col].values
        events_idx = np.where(day_df[event_col].values)[0]

        for j in events_idx:
            if j + horizon >= len(day_df):
                continue
            base_price = prices[j]
            future_prices = prices[j : j + horizon + 1]
            path = future_prices / base_price - 1.0
            paths.append(path)

    if len(paths) == 0:
        return {
            "offsets": offsets,
            "paths": np.empty((0, horizon + 1)),
            "mean_path": np.zeros(horizon + 1),
            "n_events": 0,
        }

    paths = np.vstack(paths)
    mean_path = paths.mean(axis=0)

    return {
        "offsets": offsets,
        "paths": paths,
        "mean_path": mean_path,
        "n_events": paths.shape[0],
    }


def plot_event_forward_path_multi_pretty(
    df_dict: dict,
    event_col: str,
    price_col: str = "lastPx",
    horizon: int = 30,
    stock_list=None,
):
    """
    ÂØπ df_dict ‰∏≠ÁöÑÂ§öÂè™ËÇ°Á•®ÔºåÁîª‚Äú‰∫ã‰ª∂‰πãÂêéÊú™Êù• horizon ÂàÜÈíüÂπ≥ÂùáË∑ØÂæÑ‚ÄùÔºåÊ®™ÁùÄ 3 ÂπÖÂ≠êÂõæ„ÄÇ

    ÂèÇÊï∞Ôºö
        df_dict   : {stock: DataFrame}ÔºåÊØè‰∏™ df Ëá≥Â∞ëÂåÖÂê´ event_col & price_col
        event_col : ‰∫ã‰ª∂ÂàóÂêçÔºåÂ¶Ç 'jump'„ÄÅ'spread_abnormal'
        price_col : ‰ª∑Ê†ºÂàóÂêçÔºåÂ¶Ç 'lastPx'
        horizon   : ÂêëÂêéÁúãÁöÑÂàÜÈíüÊï∞
        stock_list: Ë¶ÅÁîªÁöÑËÇ°Á•®ÂàóË°®ÔºàÊúÄÂ§ö 3 ‰∏™Ôºâ„ÄÇNone ÂàôÈªòËÆ§Âèñ df_dict Ââç 3 ‰∏™ key„ÄÇ
    """

    all_stocks = list(df_dict.keys())
    if stock_list is None:
        stock_list = sorted(all_stocks)[:3]
    else:
        stock_list = [s for s in stock_list if s in df_dict]

    n = len(stock_list)
    if n == 0:
        print("‚ö†Ô∏è No valid stocks to plot.")
        return

    # Ê†∑ÂºèËÆæÁΩÆÔºàÂíå‰Ω†ÂâçÈù¢‰øùÊåÅ‰∏ÄËá¥Ôºâ
    text_color = "#3d405b"
    panel_bg   = (244/255, 241/255, 222/255, 0.55)   # ÂçäÈÄèÊòé #f4f1de
    line_color = "#e07a5f"

    plt.style.use("ggplot")
    fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=True)
    axes = axes.flatten()

    # ËÆæÁΩÆÂ≠êÂõæËÉåÊôØ
    for ax in axes:
        ax.set_facecolor(panel_bg)

    for i, stock in enumerate(stock_list):
        ax = axes[i]
        df = df_dict[stock]

        res = event_forward_path(
            df,
            event_col=event_col,
            price_col=price_col,
            horizon=horizon,
        )

        ax.plot(
            res["offsets"],
            res["mean_path"],
            color=line_color,
            linewidth=1.8,
            label=f"{stock} (n={res['n_events']})",
        )

        ax.axhline(0, color=text_color, linestyle="--", linewidth=1, alpha=0.7)

        ax.set_title(
            f"{stock}",
            fontsize=14,
            fontweight="bold",
            color=text_color,
        )
        ax.set_xlabel("Minutes after event", color=text_color)
        if i == 0:
            ax.set_ylabel("Average return since event", color=text_color)

        ax.tick_params(axis="x", colors=text_color)
        ax.tick_params(axis="y", colors=text_color)

        ax.spines["top"].set_visible(False)
        ax.spines["right"].set_visible(False)
        ax.spines["left"].set_color(text_color)
        ax.spines["bottom"].set_color(text_color)

        ax.grid(axis="y", alpha=0.25, color=text_color)
        ax.legend(frameon=False, fontsize=9, loc="best")

    # Âà†Èô§Ê≤°Áî®Âà∞ÁöÑÂ≠êÂõæ
    for j in range(n, 3):
        fig.delaxes(axes[j])

    fig.suptitle(
        f"Average {price_col} path after {event_col}",
        fontsize=16,
        fontweight="bold",
        color=text_color,
        y=1.05,
    )

    fig.tight_layout()
    plt.show()
```

```python
def event_forward_path_by_direction(
    df: pd.DataFrame,
    event_col: str,
    ret_col: str = "px_ret",
    price_col: str = "lastPx",
    horizon: int = 30,
):
    """
    ÈÄöÁî®ÁâàÔºöÂàÜÊñπÂêë‰∫ã‰ª∂‰πãÂêéÊú™Êù•‰ª∑Ê†ºË∑ØÂæÑÔºàÂçïËÇ°Á•®Ôºâ

    ËæìÂÖ•:
        df: ÂàÜÈíüÁ∫ß DataFrameÔºåindex ‰∏∫Êó∂Èó¥Êà≥ÔºåËá≥Â∞ëÂåÖÂê´Ôºö
            - event_col  : ‰∫ã‰ª∂Ôºàbool / 0,1Ôºâ
            - ret_col    : Áî®‰∫éÂà§Êñ≠ÊñπÂêëÁöÑ‚ÄúÂΩìÊúüÊî∂Áõä‚ÄùÔºà>0 ‰∏∫ upÔºå<0 ‰∏∫ downÔºâ
            - price_col  : ‰ª∑Ê†ºÔºåÁî®‰∫éÁîªË∑ØÂæÑ
        event_col: ‰∫ã‰ª∂ÂàóÂêçÔºàÂ¶Ç 'jump', 'spread_abnormal'Ôºâ
        ret_col  : ÊñπÂêëÂàóÂêçÔºàÂ¶Ç 'px_ret'Ôºâ
        price_col: ‰ª∑Ê†ºÂàóÂêçÔºàÂ¶Ç 'lastPx'Ôºâ
        horizon  : ÂêëÂêéÁúãÁöÑÂàÜÈíüÊï∞

    ËøîÂõû:
        result: dictÔºåÂåÖÂê´ 'up' Âíå 'down' ‰∏§‰∏™Â≠ê dictÔºö
            - offsets   : [0..horizon]
            - paths     : (n_events, horizon+1)
            - mean_path : Âπ≥ÂùáË∑ØÂæÑ
            - n_events  : ‰∫ã‰ª∂‰∏™Êï∞
    """
    df = df.copy()
    if "date" not in df.columns:
        df["date"] = df.index.date

    df[event_col] = df[event_col].astype(bool)

    offsets = np.arange(horizon + 1)

    def _collect_paths(direction_sign: str):
        """
        direction_sign: 'up' -> ret_col > 0
                        'down' -> ret_col < 0
        """
        paths = []

        for d, day_df in df.groupby("date"):
            day_df = day_df.sort_index()
            prices = day_df[price_col].values
            events = day_df[event_col].values
            rets   = day_df[ret_col].values

            if direction_sign == "up":
                mask = (events) & (rets > 0)
            else:
                mask = (events) & (rets < 0)

            idx = np.where(mask)[0]
            for j in idx:
                if j + horizon >= len(day_df):
                    continue
                base_price = prices[j]
                future_prices = prices[j:j + horizon + 1]
                path = future_prices / base_price - 1.0
                paths.append(path)

        if len(paths) == 0:
            return {
                "offsets": offsets,
                "paths": np.empty((0, horizon + 1)),
                "mean_path": np.zeros(horizon + 1),
                "n_events": 0,
            }

        paths = np.vstack(paths)
        mean_path = paths.mean(axis=0)
        return {
            "offsets": offsets,
            "paths": paths,
            "mean_path": mean_path,
            "n_events": paths.shape[0],
        }

    up_res = _collect_paths("up")
    down_res = _collect_paths("down")

    return {"up": up_res, "down": down_res}


def plot_event_forward_path_by_direction_multi(
    df_dict: dict,
    event_col: str,
    ret_col: str = "px_ret",
    price_col: str = "lastPx",
    horizon: int = 30,
    stock_list=None,
):
    """
    ÂØπ df_dict ‰∏≠ÁöÑÂ§öÂè™ËÇ°Á•®ÔºåÁîª‚ÄúÂêë‰∏ä/Âêë‰∏ã‰∫ã‰ª∂‰πãÂêéÁöÑÂπ≥Âùá‰ª∑Ê†ºË∑ØÂæÑ‚ÄùÔºåÊ®™ÁùÄ‰∏âÂπÖÂ≠êÂõæ„ÄÇ

    ÂèÇÊï∞Ôºö
        df_dict   : {stock: DataFrame}
        event_col : ‰∫ã‰ª∂ÂàóÂêçÔºàÂ¶Ç 'jump'Ôºâ
        ret_col   : ÊñπÂêëÂàóÂêçÔºàÂ¶Ç 'px_ret'Ôºâ
        price_col : ‰ª∑Ê†ºÂàóÂêçÔºàÂ¶Ç 'lastPx'Ôºâ
        horizon   : ÂêëÂêéÂàÜÈíüÊï∞
        stock_list: Ë¶ÅÁîªÁöÑËÇ°Á•®‰ª£Á†ÅÂàóË°®ÔºàÊúÄÂ§ö‰∏âÂè™ÔºâÔºåÈªòËÆ§ÂèñÂ≠óÂÖ∏ÈáåÂâç 3 ‰∏™
    """
    all_stocks = list(df_dict.keys())
    if stock_list is None:
        stock_list = sorted(all_stocks)[:3]
    else:
        stock_list = [s for s in stock_list if s in df_dict]

    n = len(stock_list)
    if n == 0:
        print("‚ö†Ô∏è No valid stocks to plot.")
        return

    # Ê†∑Âºè
    text_color = "#3d405b"
    panel_bg   = (244/255, 241/255, 222/255, 0.4)   # ÂçäÈÄèÊòé #f4f1de
    up_color   = "#e07a5f"
    down_color = "#81b29a"

    plt.style.use("ggplot")
    fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=True)
    axes = axes.flatten()

    # Â≠êÂõæËÉåÊôØ
    for ax in axes:
        ax.set_facecolor(panel_bg)

    for i, stock in enumerate(stock_list):
        ax = axes[i]
        df = df_dict[stock]

        res_dir = event_forward_path_by_direction(
            df,
            event_col=event_col,
            ret_col=ret_col,
            price_col=price_col,
            horizon=horizon,
        )

        up_res = res_dir["up"]
        down_res = res_dir["down"]

        ax.plot(
            up_res["offsets"],
            up_res["mean_path"],
            label=f"Up (n={up_res['n_events']})",
            color=up_color,
            linewidth=1.8,
        )
        ax.plot(
            down_res["offsets"],
            down_res["mean_path"],
            label=f"Down (n={down_res['n_events']})",
            color=down_color,
            linewidth=1.8,
        )

        ax.axhline(0, color=text_color, linestyle="--", linewidth=1, alpha=0.7)

        ax.set_title(
            f"{stock}",
            fontsize=14,
            fontweight="bold",
            color=text_color,
        )
        ax.set_xlabel("Minutes after event", color=text_color)
        if i == 0:
            ax.set_ylabel("Average return since event", color=text_color)

        ax.tick_params(axis="x", colors=text_color)
        ax.tick_params(axis="y", colors=text_color)

        ax.spines["top"].set_visible(False)
        ax.spines["right"].set_visible(False)
        ax.spines["left"].set_color(text_color)
        ax.spines["bottom"].set_color(text_color)

        ax.grid(axis="y", alpha=0.25, color=text_color)
        ax.legend(frameon=False, fontsize=9, loc="best")

    # Â§ö‰ΩôÂ≠êÂõæÂà†Êéâ
    for j in range(n, 3):
        fig.delaxes(axes[j])

    fig.suptitle(
        f"Average {price_col} path after {event_col} (by direction)",
        fontsize=16,
        fontweight="bold",
        color=text_color,
        y=1.05,
    )

    fig.tight_layout()
    plt.show()
```

```python
def detect_abnormal_spread_rolling(
    df: pd.DataFrame,
    tick_size: float = 0.01,
    window: int = 265 * 5,   # 5 trading days of minute data
    z_thresh: float = 3.0,   # relative threshold
    abs_tick_thresh: float = 5.0  # absolute threshold in ticks
):
    """
    Rolling-based detection of abnormally wide bid-ask spreads.

    Parameters
    ----------
    df : DataFrame
        Must contain:
        - SP1: best ask price
        - BP1: best bid price
        - lastPx, date (optional: used only if you want within-day px_ret)
    tick_size : float
        Minimum tick size (e.g. 0.01).
    window : int
        Rolling window length in number of minutes.
    z_thresh : float
        Z-score threshold for relative abnormality.
    abs_tick_thresh : float
        Absolute threshold: spread must be at least this many ticks.

    Returns
    -------
    df_out : DataFrame
        With added columns:
        - spread
        - spread_ticks
        - spread_med
        - spread_std
        - spread_z
        - spread_abnormal
    """

    df = df.copy()
    df["px_ret"] = df.groupby("date")["lastPx"].diff()

    # 1. Compute current spread in price & ticks
    df["spread"] = df["SP1"] - df["BP1"]
    df["spread_ticks"] = df["spread"] / tick_size

    # 2. Rolling baseline (median + std)
    df["spread_med"] = (
        df["spread_ticks"]
        .rolling(window=window, min_periods=window//2)
        .median()
    )
    df["spread_std"] = (
        df["spread_ticks"]
        .rolling(window=window, min_periods=window//2)
        .std()
    )

    # Avoid zero std
    df["spread_std"] = df["spread_std"].replace(0, np.nan)

    # 3. Relative abnormality: z-score
    df["spread_z"] = (df["spread_ticks"] - df["spread_med"]) / df["spread_std"]

    # 4. Absolute threshold
    cond_abs = df["spread_ticks"] >= abs_tick_thresh
    cond_rel = df["spread_z"] >= z_thresh

    # 5. Final abnormal flag
    df["spread_abnormal"] = cond_abs & cond_rel

    return df
```

```python
import numpy as np
import pandas as pd

def detect_depth_collapse_rolling(
    df: pd.DataFrame,
    window: int = 265 * 5,   # 5 days of minute data
    frac_thresh: float = 0.2,   # ÂΩìÂâçÊ∑±Â∫¶ ‚â§ rolling median * 20%
    z_thresh: float = -3.0      # depth_z ‚â§ -3 => ÂºÇÂ∏∏ÂèòËñÑ
):

    df = df.copy()
    df["px_ret"] = df.groupby("date")["lastPx"].diff()

    # ----------------------------- 1. Compute total depths -----------------------------
    df["depth1_total"] = df["SV1"] + df["BV1"]

    sv_cols = [f"SV{i}" for i in range(1, 6)]
    bv_cols = [f"BV{i}" for i in range(1, 6)]
    df["depth5_total"] = df[sv_cols].sum(axis=1) + df[bv_cols].sum(axis=1)

    # ----------------------------- 2. Rolling median/std -----------------------------
    # Level 1
    df["depth1_med"] = df["depth1_total"].rolling(
        window=window, min_periods=window // 2
    ).median()

    df["depth1_std"] = df["depth1_total"].rolling(
        window=window, min_periods=window // 2
    ).std()

    # Level 5
    df["depth5_med"] = df["depth5_total"].rolling(
        window=window, min_periods=window // 2
    ).median()

    df["depth5_std"] = df["depth5_total"].rolling(
        window=window, min_periods=window // 2
    ).std()

    # Avoid zero std (which destroys z-score)
    df["depth1_std"] = df["depth1_std"].replace(0, np.nan)
    df["depth5_std"] = df["depth5_std"].replace(0, np.nan)

    # ----------------------------- 3. Ratio + z-score -----------------------------
    df["depth1_ratio"] = df["depth1_total"] / df["depth1_med"]
    df["depth5_ratio"] = df["depth5_total"] / df["depth5_med"]

    df["depth1_z"] = (df["depth1_total"] - df["depth1_med"]) / df["depth1_std"]
    df["depth5_z"] = (df["depth5_total"] - df["depth5_med"]) / df["depth5_std"]

    # ----------------------------- 4. Collapse conditions -----------------------------
    cond1_frac = df["depth1_ratio"] <= frac_thresh
    cond1_z    = df["depth1_z"] <= z_thresh
    df["depth1_collapse"] = cond1_frac & cond1_z

    cond5_frac = df["depth5_ratio"] <= frac_thresh
    cond5_z    = df["depth5_z"] <= z_thresh
    df["depth5_collapse"] = cond5_frac & cond5_z

    # Final collapse indicator
    df["depth_collapse"] = df["depth1_collapse"] | df["depth5_collapse"]

    return df
```

```python
def detect_large_trade_size_rolling(
    df,
    size_col="size",
    window=265*5,      # 5 days of minutes
    z_thresh=3.0,
    ratio_thresh=5.0,
    abs_thresh=None
):
    df = df.copy()
    df["px_ret"] = df.groupby("date")["lastPx"].diff()

    # Rolling baseline
    df["size_med"] = df[size_col].rolling(window, min_periods=window//2).median()
    df["size_std"] = df[size_col].rolling(window, min_periods=window//2).std()

    df["size_ratio"] = df[size_col] / df["size_med"]
    df["size_z"] = (df[size_col] - df["size_med"]) / df["size_std"]

    cond = (df["size_ratio"] >= ratio_thresh) & (df["size_z"] >= z_thresh)

    if abs_thresh is not None:
        cond &= (df[size_col] >= abs_thresh)

    df["large_trade"] = cond
    return df
```

```python
import numpy as np
import pandas as pd

def add_orderbook_pressure(
    df: pd.DataFrame,
    bid_prefix: str = "BV",
    ask_prefix: str = "SV",
    max_level: int = 5,
):
    """
    Âú® df ‰∏≠Âä†ÂÖ• 1-level Âíå 5-level order book pressure.

    ÈúÄË¶ÅÁöÑÂàó:
    - BV1, SV1, ..., BV5, SV5  (Â¶ÇÊûú max_level=5)
    """
    df = df.copy()

    # 1-level: (BV1 - SV1) / (BV1 + SV1)
    bv1 = df[f"{bid_prefix}1"]
    sv1 = df[f"{ask_prefix}1"]
    denom1 = bv1 + sv1
    denom1 = denom1.replace(0, np.nan)

    df["obp1"] = (bv1 - sv1) / denom1

    # 5-level: sum(BVi - SVi) / sum(BVi + SVi)
    bid_cols = [f"{bid_prefix}{i}" for i in range(1, max_level + 1)]
    ask_cols = [f"{ask_prefix}{i}" for i in range(1, max_level + 1)]

    bv_sum = df[bid_cols].sum(axis=1)
    sv_sum = df[ask_cols].sum(axis=1)
    denom5 = bv_sum + sv_sum
    denom5 = denom5.replace(0, np.nan)

    df["obp5"] = (bv_sum - sv_sum) / denom5

    return df

def detect_abnormal_pressure_rolling(
    df: pd.DataFrame,
    pressure_col: str = "obp5",
    window: int = 265 * 5,     # 5 trading days
    z_thresh: float = 3.0,     # Áõ∏ÂØπÈòàÂÄº
    abs_thresh: float = 0.5,   # ÁªùÂØπÈòàÂÄº (|pressure| >= 0.5)
):
    """
    ÂØπÁªôÂÆöÁöÑ pressure ÂàóÂÅö rolling ÂºÇÂ∏∏Ê£ÄÊµã„ÄÇ

    ËøîÂõûÂ¢ûÂä†:
    - {pressure_col}_med
    - {pressure_col}_std
    - {pressure_col}_z
    - {pressure_col}_pos_abn  (ÊûÅÁ´Ø‰π∞ÁõòÂéãÂäõ)
    - {pressure_col}_neg_abn  (ÊûÅÁ´ØÂçñÁõòÂéãÂäõ)
    - {pressure_col}_abn      (‰ªª‰∏ÄÊñπÂêëÁöÑÊûÅÁ´ØÂéãÂäõ)
    """
    df = df.copy()
    df = add_orderbook_pressure(df, bid_prefix="BV", ask_prefix="SV", max_level=5)

    # rolling Âü∫ÂáÜ
    med_col = f"{pressure_col}_med"
    std_col = f"{pressure_col}_std"
    z_col   = f"{pressure_col}_z"

    df[med_col] = (
        df[pressure_col]
        .rolling(window=window, min_periods=window // 2)
        .median()
    )
    df[std_col] = (
        df[pressure_col]
        .rolling(window=window, min_periods=window // 2)
        .std()
    )

    df[std_col] = df[std_col].replace(0, np.nan)

    df[z_col] = (df[pressure_col] - df[med_col]) / df[std_col]

    # ÁªùÂØπ + Áõ∏ÂØπ ‰∏§‰∏™Êù°‰ª∂
    cond_abs = df[pressure_col].abs() >= abs_thresh
    cond_pos = (df[pressure_col] >= abs_thresh) & (df[z_col] >= z_thresh)
    cond_neg = (df[pressure_col] <= -abs_thresh) & (df[z_col] <= -z_thresh)

    df[f"{pressure_col}_pos_abn"] = cond_pos
    df[f"{pressure_col}_neg_abn"] = cond_neg
    df[f"{pressure_col}_abn"]     = cond_abs & (df[z_col].abs() >= z_thresh)

    return df

```

```python
MODEL_RESULTS_DIR = PROJECT_ROOT / "model_results"
stock_list = ["0050", "0056", "2330"]

def load_abnormal_days_dict(model_results_dir: Path, stocks, version="raw"):
    """
    ËØªÂèñ model_results ÁõÆÂΩï‰∏ãÂêÑËÇ°Á•®ÁöÑ *_abnormal_days_xxx.csv
    ËøîÂõû: {stock: df_abnormal_days}
          df Âàó: date (datetime.date), disc_score (float), is_abnormal (0/1)
    """
    abnormal_dict = {}

    for s in stocks:
        fname = f"{s}_abnormal_days_{version}.csv"
        fpath = model_results_dir / fname
        if not fpath.exists():
            print(f"[warn] file not found: {fpath}")
            continue

        df = pd.read_csv(fpath)
        df["date"] = pd.to_datetime(df["date"]).dt.date
        df["disc_score"] = df["disc_score"].astype(float)
        df["is_abnormal"] = df["is_abnormal"].astype(int)
        abnormal_dict[s] = df

    return abnormal_dict

abnormal_days_dict = load_abnormal_days_dict(MODEL_RESULTS_DIR, stock_list, version="raw")
abnormal_days_dict["0050"].head()
```

```python
def build_disc_ic_table(
    stock_data_dict: dict,
    abnormal_days_dict: dict,
    price_col: str = "lastPx",
):
    """
    ËÆ°ÁÆó GAN ÁöÑ disc_score ‰∏éÊó•Êî∂ÁõäÁöÑÁõ∏ÂÖ≥ÊÄßÔºàIC Ë°®Ôºâ„ÄÇ

    ËæìÂÖ•Ôºö
        stock_data_dict : {stock: {"minutely_df": df, ...}}
        abnormal_days_dict : {stock: df_abn}ÔºåÊù•Ëá™ load_abnormal_days_dict
        price_col : ÂàÜÈíüÊï∞ÊçÆÈáåÁöÑ‰ª∑Ê†ºÂàóÂêçÔºàÈªòËÆ§ lastPxÔºâ

    ËæìÂá∫Ôºö
        result_df : DataFrameÔºåindex ‰∏∫ËÇ°Á•® + 'MEAN'ÔºåÂàóÔºö
            - n_days
            - mean_disc_score
            - mean_ret_today
            - mean_ret_next_day
            - ic_today      (disc_score vs ret_today)
            - ic_next       (disc_score vs ret_next_day)
    """
    rows = []

    for stock, data in stock_data_dict.items():
        if "minutely_df" not in data or stock not in abnormal_days_dict:
            print(f"[warn] skip {stock}: missing minutely_df or abnormal_days")
            continue

        # ---------- 1. ‰ªéÂàÜÈíüÊï∞ÊçÆÁÆóÊó•Êî∂Áõä ----------
        df_min = data["minutely_df"].copy()
        if "date" not in df_min.columns:
            df_min["date"] = df_min.index.date
        df_min["date"] = pd.to_datetime(df_min["date"]).dt.date

        daily_first = df_min.groupby("date")[price_col].first()
        daily_last  = df_min.groupby("date")[price_col].last()
        ret_today = (daily_last / daily_first - 1).rename("ret_today")

        daily_ret = ret_today.reset_index().sort_values("date")
        daily_ret["ret_next_day"] = daily_ret["ret_today"].shift(-1)

        # ---------- 2. Ë∑ü disc_score ÂêàÂπ∂ ----------
        df_abn = abnormal_days_dict[stock].copy()  # Êúâ date, disc_score, is_abnormal
        df_abn["date"] = pd.to_datetime(df_abn["date"]).dt.date

        merged = pd.merge(df_abn, daily_ret, on="date", how="inner")

        # ‰∏¢ÊéâÊ≤°Êúâ next_day ÁöÑÔºàÊúÄÂêé‰∏Ä‰∏™Êó•ÊúüÔºâ
        merged = merged.dropna(subset=["ret_today", "ret_next_day", "disc_score"])

        if merged.shape[0] < 2:
            ic_today = np.nan
            ic_next  = np.nan
        else:
            ic_today = merged["disc_score"].corr(merged["ret_today"])
            ic_next  = merged["disc_score"].corr(merged["ret_next_day"])

        row = {
            "stock": stock,
            "n_days": int(merged.shape[0]),
            "mean_disc_score": float(merged["disc_score"].mean()),
            "mean_ret_today": float(merged["ret_today"].mean()),
            "mean_ret_next_day": float(merged["ret_next_day"].mean()),
            "ic_today": float(ic_today) if np.isfinite(ic_today) else np.nan,
            "ic_next": float(ic_next) if np.isfinite(ic_next) else np.nan,
        }
        rows.append(row)

    result_df = pd.DataFrame(rows).set_index("stock")

    # ---------- 3. Âä†‰∏Ä‰∏™ÂùáÂÄºË°å ----------
    mean_row = {
        "stock": "MEAN",
        "n_days": np.nan,
        "mean_disc_score": result_df["mean_disc_score"].mean(),
        "mean_ret_today": result_df["mean_ret_today"].mean(),
        "mean_ret_next_day": result_df["mean_ret_next_day"].mean(),
        "ic_today": result_df["ic_today"].mean(),
        "ic_next": result_df["ic_next"].mean(),
    }
    result_df = pd.concat(
        [result_df, pd.DataFrame([mean_row]).set_index("stock")]
    )

    return result_df
```

```python
def convert_minutely_to_daily_counts(df_min):
    """
    ËæìÂÖ•: ÂàÜÈíüÁ∫ß dataframe (Âê´‰∏Ä‰∏™ÊàñÂ§ö‰∏™ 0/1 ÂºÇÂ∏∏Âàó)
    ËæìÂá∫: Êó•È¢ë‰∫ã‰ª∂ËÆ°Êï∞ (date, event_1, event_2, ...)
    """
    df = df_min.copy()

    if "date" not in df.columns:
        df["date"] = df.index.date
    df["date"] = pd.to_datetime(df["date"]).dt.date

    # Ëá™Âä®ËØÜÂà´‰∫ã‰ª∂ÂàóÔºöÊâÄÊúâ int/bool ÂàóËßÜ‰∏∫‰∫ã‰ª∂
    event_cols = [
        col for col in df.columns
        if df[col].dropna().isin([0,1]).all()
    ]

    # ÁªüËÆ°ÊØèÂ§©ÁöÑ‰∫ã‰ª∂Êï∞Èáè
    daily = df.groupby("date")[event_cols].sum().reset_index()

    return daily

def build_daily_manual_dict(event_dicts, stocks):
    """
    event_dicts: ÊâÄÊúâ‰∫ã‰ª∂ÁöÑ dict ÁöÑ list
                 [jump_dict, spread_dict, size_dict, depth_dict,
                  pressure_1_dict, pressure_5_dict]

    ËæìÂá∫:
       { stock : dataframe(date, event1, event2, ...) }
       Ëá™Âä®ÂØπÂêå‰∏ÄÂ§©ÁöÑ‰∫ã‰ª∂ËøõË°åÂêàÂπ∂Ôºàinner mergeÔºâ
    """
    daily_dict = {}

    for stock in stocks:
        merged = None
        for d in event_dicts:
            if stock not in d:
                continue
            df_min = d[stock]
            df_daily = convert_minutely_to_daily_counts(df_min)

            if merged is None:
                merged = df_daily
            else:
                merged = pd.merge(merged, df_daily, on="date", how="outer")

        if merged is not None:
            merged = merged.fillna(0)
            daily_dict[stock] = merged

    return daily_dict

def build_daily_price_df(minutely_df, price_col="lastPx"):
    """
    ‰ªé minute-level Êï∞ÊçÆÊèêÂèñ daily open, close, return„ÄÇ
    """
    df = minutely_df.copy()

    if "date" not in df.columns:
        df["date"] = df.index.date
    df["date"] = pd.to_datetime(df["date"]).dt.date

    daily_first = df.groupby("date")[price_col].first().rename("open")
    daily_last  = df.groupby("date")[price_col].last().rename("close")

    daily = pd.concat([daily_first, daily_last], axis=1).reset_index()
    daily["ret_today"] = daily["close"] / daily["open"] - 1
    daily["ret_next"]  = daily["ret_today"].shift(-1)

    return daily

def merge_all_for_analysis(
    stock_data_dict,
    abnormal_days_dict,
    daily_manual_dict,
    stocks,
    price_col="lastPx",
):
    """
    ËæìÂá∫: merged_analysis_dict
    ÊØèÂè™ËÇ°Á•®:
        date, disc_score, is_abnormal,
        open, close, ret_today, ret_next,
        + ÊâÄÊúâ manual event counts
    """
    merged_dict = {}

    for stock in stocks:

        if stock not in abnormal_days_dict:
            print(f"[warn] no GAN result for {stock}")
            continue
        
        # ---------------- 1. GAN Êó•È¢ë ----------------
        df_gan = abnormal_days_dict[stock].copy()

        # ---------------- 2. Êó•È¢ë‰ª∑Ê†º ----------------
        if stock not in stock_data_dict:
            print(f"[warn] no minute data for {stock}")
            continue

        df_min = stock_data_dict[stock]["minutely_df"]
        df_price = build_daily_price_df(df_min, price_col=price_col)

        # merge ‰ª∑Ê†º
        merged = pd.merge(df_gan, df_price, on="date", how="left")

        # ---------------- 3. ‰∫∫Â∑•‰∫ã‰ª∂Êó•È¢ë ----------------
        df_manual = daily_manual_dict.get(stock)
        if df_manual is not None:
            merged = pd.merge(merged, df_manual, on="date", how="left")

        merged = merged.fillna(0)

        merged_dict[stock] = merged

    return merged_dict


stocks = ["0050", "0056", "2330"]

event_dicts = [
    jump_dict,
    spread_dict,
    size_dict,
    collapse_dict,        # depth collapse
    pressure_1_dict,
    pressure_5_dict,
]

daily_manual_dict = build_daily_manual_dict(event_dicts, stocks)

merged_analysis_dict = merge_all_for_analysis(
    stock_data_dict=stock_data_dict,
    abnormal_days_dict=abnormal_days_dict,
    daily_manual_dict=daily_manual_dict,
    stocks=stocks,
    price_col="lastPx",
)

```

```python
import numpy as np
import pandas as pd
from scipy import stats

def compute_univariate_tests_for_stock(
    df: pd.DataFrame,
    abnormal_cols: list,
    delta: int = 1,
):
    """
    ÂØπÂçïÂè™ËÇ°Á•®ÂÅö abnormal vs normal ÁöÑ return univariate testÔºö
    - mean_abn, mean_norm
    - diff
    - t-test (Welch)
    - KS test

    ÂèÇÊï∞
    ----
    df : ÂêàÂπ∂ÂêéÊó•È¢ë dataframeÔºàÂåÖÂê´ disc_score, is_abnormal, ÊâÄÊúâÊâãÂ∑•‰∫ã‰ª∂Ôºâ
    abnormal_cols : Ë¶ÅÊ£ÄÈ™åÁöÑÂàóÔºåÊØîÂ¶Ç ["is_abnormal", "jump", "spread_abnormal", ...]
    delta : return horizonÔºå‰æãÂ¶Ç 1 = ‰∏ã‰∏ÄÂàÜÈíüÔºåËã•‰Ω†ÁöÑ return Â∑≤ÁªèÊòØÊó•È¢ëÔºåÂàô delta=1 Âç≥ next-day return„ÄÇ

    ËøîÂõû
    ----
    DataFrameÔºåÊØèË°åÂØπÂ∫î‰∏Ä‰∏™‰∫ã‰ª∂
    """

    df = df.copy().sort_values("date")

    results = []

    for col in abnormal_cols:
        if col not in df.columns:
            print(f"[warn] skip {col}: not in df")
            continue

        A = df[col].astype(bool)
        r = df["ret_next"]

        r_abn  = r[A].dropna()
        r_norm = r[~A].dropna()

        n_abn, n_norm = len(r_abn), len(r_norm)

        if n_abn < 2 or n_norm < 2:
            mean_abn = r_abn.mean() if n_abn > 0 else np.nan
            mean_norm = r_norm.mean() if n_norm > 0 else np.nan
            t_stat = t_pval = ks_stat = ks_pval = np.nan
        else:
            mean_abn  = r_abn.mean()
            mean_norm = r_norm.mean()

            t_stat, t_pval = stats.ttest_ind(
                r_abn, r_norm, equal_var=False, nan_policy="omit"
            )

            ks_stat, ks_pval = stats.ks_2samp(
                r_abn, r_norm, alternative="two-sided"
            )

        results.append({
            "abnormal_col": col,
            "n_abn": n_abn,
            "n_norm": n_norm,
            "mean_abn": mean_abn,
            "mean_norm": mean_norm,
            "diff": mean_abn - mean_norm,
            "t_stat": t_stat,
            "t_pvalue": t_pval,
            "ks_stat": ks_stat,
            "ks_pvalue": ks_pval,
        })

    return pd.DataFrame(results).set_index("abnormal_col")

def build_univariate_test_table(
    merged_analysis_dict: dict,
    abnormal_cols: list,
    delta: int = 1,
):
    """
    ÂØπÂ§öÂè™ËÇ°Á•®ÂÅö univariate t-test & KS test„ÄÇ
    
    merged_analysis_dict:
        { stock: merged_df }, merged_df Êù•Ëá™ merge_manual_with_gan_all

    abnormal_cols:
        ÊâÄÊúâË¶ÅÊ£ÄÈ™åÁöÑ‰∫ã‰ª∂ÂàóÔºå‰æãÂ¶ÇÔºö
        ["is_abnormal", "jump", "spread_abnormal",
         "depth_collapse", "large_trade", "obp1_abn", "obp5_abn"]

    ËøîÂõû:
        MultiIndex DataFrame:
            index = (stock, abnormal_col)
            columns = n_abn, n_norm, mean_abn, mean_norm, diff,
                      t_stat, t_pvalue, ks_stat, ks_pvalue
    """
    rows = []

    for stock, df in merged_analysis_dict.items():
        stat_df = compute_univariate_tests_for_stock(
            df=df,
            abnormal_cols=abnormal_cols,
            delta=delta,
        )

        for event, row in stat_df.iterrows():
            rows.append({
                "stock": stock,
                "abnormal_col": event,
                **row.to_dict()
            })

    result_df = pd.DataFrame(rows).set_index(["stock", "abnormal_col"])
    return result_df

abnormal_cols = [
    "is_abnormal",         # GAN abnormal label
    "jump",
    "spread_abnormal",
    "depth_collapse",
    "large_trade",
    "obp1_abn",
    "obp5_abn",
]

uni_table = build_univariate_test_table(
    merged_analysis_dict=merged_analysis_dict,
    abnormal_cols=abnormal_cols,
    delta=1,       # Áî® r_{t+1}
)

```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

def run_predictive_regression_for_stock(
    df: pd.DataFrame,
    feature_cols: list,
    ret_col: str = "ret_next",
):
    """
    ÂØπÂçïÂè™ËÇ°Á•®ÂÅöÂõûÂΩí:
        ret_next = alpha + beta' * X + eps

    ÂèÇÊï∞
    ----
    df : merged_analysis_dict[stock] ÂØπÂ∫îÁöÑ DataFrame
         ÈúÄË¶ÅËá≥Â∞ëÂåÖÂê´ ret_next Âíå feature_cols
    feature_cols : Ëá™ÂèòÈáèÂàóÂêçÂàóË°®
    ret_col : Âõ†ÂèòÈáèÂàóÂêçÔºåÈªòËÆ§ 'ret_next'

    ËøîÂõû
    ----
    result_df : DataFrame, index ‰∏∫ feature ÂêçÔºàÂê´ interceptÔºâÔºåÂàó‰∏∫ coef, t_stat, p_value
    """

    data = df.copy()

    # Âè™‰øùÁïôÈúÄË¶ÅÁöÑÂàóÂπ∂ÂéªÊéâÁº∫Â§±
    cols = [ret_col] + feature_cols
    data = data[cols].dropna()

    # ‰∏¢ÊéâÊ†∑Êú¨Êï∞Â§™Â∞ëÁöÑÊÉÖÂÜµ
    if data.shape[0] < len(feature_cols) + 5:
        print("[warn] not enough observations")
        return None

    # ÂéªÊéâÂú®ËØ•ËÇ°Á•®‰∏äÊñπÂ∑Æ‰∏∫ 0 ÁöÑ featureÔºà‰æãÂ¶ÇÊüê‰∫ã‰ª∂‰ªéÊú™ÂèëÁîüÔºâ
    effective_features = []
    for c in feature_cols:
        if data[c].std() > 0:
            effective_features.append(c)

    if len(effective_features) == 0:
        print("[warn] all features have zero variance")
        return None

    y = data[ret_col].values
    X = data[effective_features]

    # ÂèØ‰ª•ÈÄâÊã©ÂØπËÆ°Êï∞ÂÅöÊ†áÂáÜÂåñÔºàÊñπ‰æøÊØîËæÉÁ≥ªÊï∞Â§ßÂ∞èÔºâÔºõÂÖàÁªô‰Ω†ÂéüÂßãÁâàÊú¨
    # X = (X - X.mean()) / X.std(ddof=0)

    X = sm.add_constant(X)  # Âä†Êà™Ë∑ù
    model = sm.OLS(y, X)
    # Áî® heteroskedasticity-robust ÁöÑ White Ê†áÂáÜËØØ
    res = model.fit(cov_type="HC1")

    coef  = res.params
    tvals = res.tvalues
    pvals = res.pvalues

    result_df = pd.DataFrame({
        "coef": coef,
        "t_stat": tvals,
        "p_value": pvals,
    })
    result_df.index.name = "factor"

    return result_df

def run_predictive_regressions_all(
    merged_analysis_dict: dict,
    feature_cols: list,
    ret_col: str = "ret_next",
):
    """
    ÂØπÂ§öÂè™ËÇ°Á•®ÂàÜÂà´ÂõûÂΩíÔºåÊ±áÊÄªÁªìÊûú„ÄÇ

    ËøîÂõû:
        MultiIndex DataFrame:
            index = (stock, factor)
            columns = coef, t_stat, p_value
    """
    rows = []

    for stock, df in merged_analysis_dict.items():
        res_df = run_predictive_regression_for_stock(
            df=df,
            feature_cols=feature_cols,
            ret_col=ret_col,
        )
        if res_df is None:
            continue

        for factor, row in res_df.iterrows():
            rows.append({
                "stock": stock,
                "factor": factor,
                "coef": row["coef"],
                "t_stat": row["t_stat"],
                "p_value": row["p_value"],
            })

    result = pd.DataFrame(rows).set_index(["stock", "factor"]).sort_index()
    return result
```

```python
import numpy as np
import pandas as pd

def _prepare_disc_ret_for_stock(
    stock: str,
    stock_data_dict: dict,
    abnormal_days_dict: dict,
    price_col: str = "lastPx",
):
    """
    ËøîÂõûÊüêÂè™ËÇ°Á•®ÁöÑÊó•Â∫¶Ë°®Ôºö
        date, disc_score, ret_today, ret_next_day
    """
    if stock not in stock_data_dict or "minutely_df" not in stock_data_dict[stock]:
        raise ValueError(f"{stock}: missing minutely_df in stock_data_dict")
    if stock not in abnormal_days_dict:
        raise ValueError(f"{stock}: missing disc_score info in abnormal_days_dict")

    df_min = stock_data_dict[stock]["minutely_df"].copy()
    if "date" not in df_min.columns:
        df_min["date"] = df_min.index.date
    df_min["date"] = pd.to_datetime(df_min["date"]).dt.date

    # Êó•Â∫¶Êî∂Áõä
    daily_first = df_min.groupby("date")[price_col].first()
    daily_last  = df_min.groupby("date")[price_col].last()
    ret_today   = (daily_last / daily_first - 1).rename("ret_today")

    daily_ret = ret_today.reset_index().sort_values("date")
    daily_ret["ret_next_day"] = daily_ret["ret_today"].shift(-1)

    # merge disc_score
    df_abn = abnormal_days_dict[stock].copy()  # Êúâ date, disc_score
    df_abn["date"] = pd.to_datetime(df_abn["date"]).dt.date

    merged = pd.merge(df_abn, daily_ret, on="date", how="inner")
    merged = merged.dropna(subset=["disc_score", "ret_today", "ret_next_day"])

    return merged

import matplotlib.pyplot as plt

def plot_disc_vs_return_multi(
    stock_data_dict: dict,
    abnormal_days_dict: dict,
    price_col: str = "lastPx",
    ret_horizon: str = "today",   # "today" or "next"
    stock_list=None,
):
    """
    Áîª‰∏âÂè™ËÇ°Á•®ÁöÑÊï£ÁÇπÂõæÔºö
        x: disc_score
        y: Êó•Êî∂ÁõäÔºà‰ªäÂ§©ÊàñÊòéÂ§©Ôºâ
    ‰∏â‰∏™Â≠êÂõæÊ®™ÊéíÔºåÈ£éÊ†ºÂíå‰πãÂâç‰∏ÄËá¥„ÄÇ
    """
    all_stocks = list(abnormal_days_dict.keys())
    if stock_list is None:
        stock_list = sorted(all_stocks)[:3]
    else:
        stock_list = [s for s in stock_list if s in abnormal_days_dict]

    n = len(stock_list)
    if n == 0:
        print("‚ö†Ô∏è No valid stocks to plot.")
        return

    if ret_horizon == "today":
        y_col = "ret_today"
        y_label = "Return (today)"
    elif ret_horizon == "next":
        y_col = "ret_next_day"
        y_label = "Return (next day)"
    else:
        raise ValueError("ret_horizon must be 'today' or 'next'")

    # Ê†∑Âºè
    text_color = "#3d405b"
    panel_bg   = (244/255, 241/255, 222/255, 0.4)   # ÂçäÈÄèÊòé #f4f1de
    scatter_color = "#e07a5f"

    plt.style.use("ggplot")
    fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=True)
    axes = axes.flatten()

    for ax in axes:
        ax.set_facecolor(panel_bg)

    for i, stock in enumerate(stock_list):
        ax = axes[i]

        merged = _prepare_disc_ret_for_stock(
            stock,
            stock_data_dict=stock_data_dict,
            abnormal_days_dict=abnormal_days_dict,
            price_col=price_col,
        )

        x = merged["disc_score"].values
        y = merged[y_col].values

        # Pearson r
        r = np.corrcoef(x, y)[0, 1] if len(x) > 1 else np.nan

        ax.scatter(x, y, s=18, alpha=0.7, color=scatter_color, edgecolors="none")

        ax.set_title(
            f"{stock} (r={r:.3f})",
            fontsize=14,
            fontweight="bold",
            color=text_color,
        )
        ax.set_xlabel("disc_score", color=text_color)
        if i == 0:
            ax.set_ylabel(y_label, color=text_color)

        ax.tick_params(axis="x", colors=text_color)
        ax.tick_params(axis="y", colors=text_color)

        ax.spines["top"].set_visible(False)
        ax.spines["right"].set_visible(False)
        ax.spines["left"].set_color(text_color)
        ax.spines["bottom"].set_color(text_color)

        ax.grid(axis="y", alpha=0.25, color=text_color)

    # Âà†Èô§Â§ö‰ΩôÂ≠êÂõæ
    for j in range(n, 3):
        fig.delaxes(axes[j])

    fig.suptitle(
        f"disc_score vs {y_label}",
        fontsize=16,
        fontweight="bold",
        color=text_color,
        y=1.05,
    )

    fig.tight_layout()
    plt.show()

# ‰ªäÂ§©Êî∂Áõä vs disc_score
plot_disc_vs_return_multi(
    stock_data_dict,
    abnormal_days_dict,
    price_col="lastPx",
    ret_horizon="today",
    stock_list=["0050", "0056", "2330"],
)

# ÊòéÂ§©Êî∂Áõä vs disc_score
plot_disc_vs_return_multi(
    stock_data_dict,
    abnormal_days_dict,
    price_col="lastPx",
    ret_horizon="next",
    stock_list=["0050", "0056", "2330"],
)
```

## ÂàÜÊûê‰∫∫‰∏∫ËßÑÂÆöÁöÑÂºÇÂ∏∏Áé∞Ë±°‰∏éGANÈ¢ÑÊµã‰πãÈó¥ÁöÑÂÖ≥Á≥ª

```python
import numpy as np
import pandas as pd

def daily_event_count(df, event_col):
    """
    ËæìÂÖ•ÔºöÂàÜÈíüÁ∫ß dfÔºàÂê´ event_col Âíå 'date'Ôºâ
    ËæìÂá∫ÔºöÊØèÂ§© event Ê¨°Êï∞ÔºàSeriesÔºåindex ‰∏∫ dateÔºâ
    """
    df = df.copy()
    if "date" not in df.columns:
        df["date"] = df.index.date
    df["date"] = pd.to_datetime(df["date"]).dt.date
    return df.groupby("date")[event_col].sum()

def build_event_disc_corr_table(
    jump_dict: dict,
    spread_dict: dict,
    collapse_dict: dict,
    size_dict: dict,                # ‚≠ê Êñ∞Â¢û
    abnormal_days_dict: dict,
    stocks=None,
):
    """
    ËÆ°ÁÆó 4 ÁßçÂºÇÂ∏∏ÔºàÊó•Â∫¶ event Ê¨°Êï∞Ôºâ‰∏é disc_score ÁöÑ Pearson Áõ∏ÂÖ≥Á≥ªÊï∞Ë°®„ÄÇ

    ‰∫ã‰ª∂:
        - jump
        - spread_abnormal
        - depth_collapse
        - large_trade    (Êù•Ëá™ size_dict)

    ËøîÂõû DataFrame:
        index : stock
        columns : ‰∏äËø∞ 4 Áßç‰∫ã‰ª∂
    """
    if stocks is None:
        s1 = set(jump_dict.keys())
        s2 = set(spread_dict.keys())
        s3 = set(collapse_dict.keys())
        s4 = set(size_dict.keys())
        s5 = set(abnormal_days_dict.keys())
        stocks = sorted(list(s1 & s2 & s3 & s4 & s5))

    # ÊâÄÊúâ‰∫ã‰ª∂Êù•Ê∫ê
    events = {
        "price_jump":            (jump_dict,     "jump"),
        "spread_abnormal": (spread_dict,   "spread_abnormal"),
        "depth_collapse":  (collapse_dict, "depth_collapse"),
        "large_trade":     (size_dict,     "large_trade"),     # ‚≠ê Êñ∞Â¢û
        "obp1_abn":        (pressure_1_dict,    "obp1_abn"),
        "obp5_abn":        (pressure_5_dict,    "obp5_abn"),
        # "obp1_pos_abn":  (pressure_1_dict, "obp1_pos_abn"),
        # "obp1_neg_abn":  (pressure_1_dict, "obp1_neg_abn"),
        # "obp5_pos_abn":  (pressure_5_dict, "obp5_pos_abn"),
        # "obp5_neg_abn":  (pressure_5_dict, "obp5_neg_abn"),
        
    }

    rows = []

    for stock in stocks:
        df_abn = abnormal_days_dict[stock].copy()
        df_abn["date"] = pd.to_datetime(df_abn["date"]).dt.date
        disc = df_abn.set_index("date")["disc_score"].astype(float)

        row = {"stock": stock}

        # ÊØèÁßç‰∫ã‰ª∂ÂàÜÂà´ÁÆó corr
        for ev_name, (dct, col) in events.items():
            if stock not in dct:
                row[ev_name] = np.nan
                continue

            # Êó•È¢ë event_count
            daily_ev = daily_event_count(dct[stock], col)

            # ÂêàÂπ∂
            merged = pd.concat(
                [daily_ev.rename("event_count"), disc.rename("disc_score")],
                axis=1,
                join="inner",
            ).dropna()

            if merged.shape[0] < 2:
                corr = np.nan
            else:
                corr = merged["event_count"].corr(merged["disc_score"])

            row[ev_name] = corr

        rows.append(row)

    result = pd.DataFrame(rows).set_index("stock")

    # Ê∑ªÂä†Âπ≥ÂùáË°å
    mean_row = pd.DataFrame(result.mean(axis=0)).T
    mean_row.index = ["MEAN"]
    result = pd.concat([result, mean_row])

    return result
```

```python
stock_list = ["0050", "0056", "2330"]

event_disc_corr_table = build_event_disc_corr_table(
    jump_dict=jump_dict,
    spread_dict=spread_dict,
    collapse_dict=collapse_dict,
    size_dict=size_dict,
    abnormal_days_dict=abnormal_days_dict,
    stocks=stock_list,
)

event_disc_corr_table
```

üß† 5. Â∫èÂàóÁªìÊûÑÊú¨Ë∫´ÂºÇÂ∏∏ÔºàTime Series ShapeÔºâ

GAN ÁöÑÂà§Âà´Âô®ÂÖ∂ÂÆûÂ∞±Âú®ÊäìËøô‰∫õÔºö

üü£Ôºà1ÔºâÊüê‰∏ÄÊó•ÁöÑ LOB Ë°åÁ®ã‰∏éËÆ≠ÁªÉÂàÜÂ∏É‰∏ç‰∏ÄËá¥

Ë°®Áé∞‰∏∫Ôºö

Ê≥¢Âä®ÁéáÁâπÂæÅÂÆåÂÖ®‰∏çÂêå

Depth ÁöÑ autocorrelation ÂºÇÂ∏∏Êñ≠Ë£Ç

spread Â∫èÂàóÂΩ¢Áä∂ÊÄ™ÂºÇ

LOB pressure Â∫èÂàóÂëàÁé∞ÈùûÊ≠£Â∏∏ÈúáËç°

ËøôÁ±ªÂ±û‰∫é Êó∂Â∫èÂΩ¢ÊÄÅÂºÇÂ∏∏Ôºàshape anomalyÔºâ
ÊâÄ‰ª• GAN ÂæàÊìÖÈïøÊäì„ÄÇ

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA

def compute_shape_features(
    minutely_df,
    features=("lastPx", "spread", "depth5"),
    lags=(1,5,10),
    fft_use=True,
    pca_use=True,
):
    """
    ÁªôÊüê‰∏ÄÂ§©ÁöÑ minutely_dfÔºà265 Ë°åÔºâËÆ°ÁÆóÊó∂Â∫èÂΩ¢Áä∂ÁâπÂæÅÔºö
    - autocorrÔºàÂêÑ lagÔºâ
    - frequency features (FFT)
    - pca reconstruction error
    
    ËøîÂõû dict
    """
    out = {}

    df = minutely_df.copy()
    df["ret"] = df["lastPx"].pct_change()
    df["spread"] = df["SP1"] - df["BP1"]
    df["depth5"] = df[[f"SV{i}" for i in range(1,6)]].sum(axis=1) + \
                   df[[f"BV{i}" for i in range(1,6)]].sum(axis=1)

    # ============= 1. Autocorr =============
    for col in features:
        x = df[col].values
        for lag in lags:
            if len(x) > lag:
                ac = np.corrcoef(x[:-lag], x[lag:])[0,1]
            else:
                ac = np.nan
            out[f"{col}_autocorr_lag{lag}"] = ac

    # ============= 2. FFT shape features =============
    if fft_use:
        for col in features:
            x = df[col].ffill().bfill().values
            freq = np.abs(np.fft.rfft(x))

            out[f"{col}_fft_energy_high"] = freq[int(len(freq)*0.5):].mean()
            out[f"{col}_fft_energy_low"]  = freq[:int(len(freq)*0.2)].mean()
            out[f"{col}_fft_total_energy"] = freq.mean()

    # ============= 3. PCA reconstruction error =============
    if pca_use:
        X = df[["lastPx","spread","depth5"]].values
        X = (X - X.mean()) / (X.std() + 1e-9)

        pca = PCA(n_components=2)
        pca.fit(X)
        X_rec = pca.inverse_transform(pca.transform(X))
        out["pca_recon_error"] = np.linalg.norm(X - X_rec)

    return out

import numpy as np
import pandas as pd

def daily_return_from_minutely(minutely_df: pd.DataFrame,
                               price_col: str = "lastPx") -> pd.Series:
    """
    Ê†πÊçÆÂàÜÈíüÁ∫ßÊï∞ÊçÆÁÆóÊó•Êî∂ÁõäÔºö
        ret_today = close / open - 1

    ËøîÂõûÔºö
        SeriesÔºåindex ‰∏∫ dateÔºàdatetime.dateÔºâÔºåname='ret_today'
    """
    df = minutely_df.copy()
    if "date" not in df.columns:
        df["date"] = df.index.date
    df["date"] = pd.to_datetime(df["date"]).dt.date

    daily_first = df.groupby("date")[price_col].first()
    daily_last  = df.groupby("date")[price_col].last()

    ret_today = (daily_last / daily_first - 1.0).rename("ret_today")
    return ret_today


def compute_daily_shape_scores(minutely_df):
    shape_list = []

    for date, day_df in minutely_df.groupby("date"):
        feats = compute_shape_features(day_df)
        feats["date"] = date
        shape_list.append(feats)

    return pd.DataFrame(shape_list).set_index("date")

def shape_factor_ic_table(
    shape_dict: dict,          
    stock_data_dict: dict,
    abnormal_days_dict: dict,
    stock_list=None,
):
    if stock_list is None:
        s1 = set(shape_dict.keys())
        s2 = set(stock_data_dict.keys())
        s3 = set(abnormal_days_dict.keys())
        stock_list = sorted(list(s1 & s2 & s3))

    rows = []

    for stock in stock_list:
        if stock not in shape_dict: continue
        if stock not in stock_data_dict: continue
        if stock not in abnormal_days_dict: continue

        shape_df = shape_dict[stock].copy()
        shape_df.index = pd.to_datetime(shape_df.index).date

        minutely_df = stock_data_dict[stock]["minutely_df"]
        ret_today = daily_return_from_minutely(minutely_df, price_col="lastPx")
        ret_next  = ret_today.shift(-1).rename("ret_next_day")

        abn = abnormal_days_dict[stock].copy()
        abn["date"] = pd.to_datetime(abn["date"]).dt.date
        disc = abn.set_index("date")["disc_score"].astype(float)

        base = shape_df.join(ret_today, how="inner")
        base = base.join(ret_next, how="inner")
        base = base.join(disc.rename("disc_score"), how="inner")

        factor_cols = [c for c in shape_df.columns if c not in ["date"]]

        for fac in factor_cols:
            tmp = base[[fac, "ret_today", "ret_next_day", "disc_score"]].dropna()
            n = tmp.shape[0]
            if n < 5:
                ic_today = ic_next = ic_disc = np.nan
            else:
                ic_today = tmp[fac].corr(tmp["ret_today"])
                ic_next  = tmp[fac].corr(tmp["ret_next_day"])
                ic_disc  = tmp[fac].corr(tmp["disc_score"])

            rows.append({
                "factor":  fac,
                "stock":   stock,
                "n_obs":   n,
                "ic_today": ic_today,
                "ic_next":  ic_next,
                "ic_disc":  ic_disc,
            })

    result = pd.DataFrame(rows)

    # ‚≠ê Êää factor ÊîæÁ¨¨‰∏ÄÂàóÔºåstock Á¨¨‰∫åÂàó
    if not result.empty:
        result = result.sort_values(["factor", "stock"])
        result = result.set_index(["factor", "stock"])

    return result

shape_dict = {}

for stock, data in stock_data_dict.items():
    df = data["minutely_df"]
    shape_scores = compute_daily_shape_scores(df)
    shape_dict[stock] = shape_scores
    

stock_list = ["0050", "0056", "2330"]

shape_ic = shape_factor_ic_table(
    shape_dict=shape_dict,
    stock_data_dict=stock_data_dict,
    abnormal_days_dict=abnormal_days_dict,
    stock_list=stock_list,
)

shape_ic
```
